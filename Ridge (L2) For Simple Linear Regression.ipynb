```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from sklearn.datasets import make_regression
```

```python
X,y = make_regression(n_samples=100 , n_features=1 , n_informative=1 , n_targets=1 , noise=20 , random_state=13)
```

```python
plt.figure(figsize=(6,4))
plt.scatter(X,y);
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X,y)
print(lr.coef_ ,'\n' , lr.intercept_)
```

```python
from sklearn.linear_model import Ridge
r = Ridge(alpha=10)
r.fit(X,y)
print(r.coef_ , '\n' , r.intercept_)
```

- __As we can see by setting __Hyperparameter(Lambda(alpha=10))__ Ridge regression Value is close to original Linear regression model__

```python
r1 = Ridge(alpha=100)
r1.fit(X,y)
print(r1.coef_ , '\n' , r1.intercept_)
```

```python
#plotting the graph

plt.plot(X,y , 'b.')
plt.plot(X,lr.predict(X) , color='red' , label='alpha=0')
plt.plot(X,r.predict(X) , color='green' , label='alpha=10')
plt.plot(X,r1.predict(X) , color='Orange' , label='alpha=100')
plt.legend();
```
